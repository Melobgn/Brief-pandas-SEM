```python
import matplotlib.pyplot as plt

# Assurez-vous que 'protein_supply_quantity_g' est en kilogrammes

df = df.mask(df["year"] == 2018).dropna()

df['protein_supply_quantity_g'] = df['protein_supply_quantity_g'] * 1000

# Top 10 des pays avec la plus grande quantité de protéines
dispo_nourriture_totale_prot_top = (
    df.groupby(['country', 'year'])[['protein_supply_quantity_g']]
      .agg(sum)
      .reset_index()
      .sort_values(['year', 'protein_supply_quantity_g'], ascending=False)
      .rename(columns={'country': 'pays', 'year': 'année', 'protein_supply_quantity_g': 'kg de protéines'})
      .head(10)
)

# Bottom 10 des pays avec la plus petite quantité de protéines
dispo_nourriture_totale_prot_bottom = (
    df.groupby(['country', 'year'])[['protein_supply_quantity_g']]
      .agg(sum)
      .reset_index()
      .sort_values(['year', 'protein_supply_quantity_g'], ascending=True)
      .rename(columns={'country': 'pays', 'year': 'année', 'protein_supply_quantity_g': 'kg de protéines'})
      .head(10)
)

# Création du graphique en barres
plt.figure(figsize=(10, 6))
fig, ax1 = plt.subplots()

color_top = 'tab:brown'
ax1.set_xlabel('Pays')
ax1.set_ylabel('Disponibilité Top 10', color=color_top)
ax1.bar(dispo_nourriture_totale_prot_top['pays'], dispo_nourriture_totale_prot_top['kg de protéines'], color=color_top)
ax1.tick_params(axis='y', labelcolor=color_top, labelsize=7)
plt.xticks(rotation=50, ha="right")

ax2 = ax1.twinx()

color_bottom = 'tab:pink'
ax2.set_ylabel('Disponibilité Bottom 10', color=color_bottom)
ax2.bar(dispo_nourriture_totale_prot_bottom['pays'], dispo_nourriture_totale_prot_bottom['kg de protéines'], color=color_bottom)
ax2.tick_params(axis='y', labelcolor=color_bottom)

plt.xticks(rotation=50, ha="right")
plt.suptitle('Comparaison de la disponibilité en protéines (kg) par an et par pays (Top 10 et Bottom 10)')

fig.tight_layout()
plt.show()
```
