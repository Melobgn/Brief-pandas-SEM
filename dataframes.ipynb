```python
# Exercice 31
# Avec Matplotlib, tracez l'évolution de la population (pop_1000_hab) de 2018 à 2019 pour les 5 pays les plus peuplés. 
# Utilisez un graphique en barres pour chaque pays pour comparer les deux années.

df_31 = df[['country','year','pop_1000_hab']]
df_31 = df_31.groupby(['country', 'year'])['pop_1000_hab'].agg(max)
df_31 = df_31.reset_index()
df_31 = df_31.sort_values('pop_1000_hab', ascending=False).head(10)
df_31

piv = df_31.pivot_table(values='pop_1000_hab', columns=['year'], index='country')
piv = piv.reset_index()
piv = piv.sort_values(by=2019, ascending=False)

piv
piv.plot(kind='bar', x="country", color=['green', 'orange'])
```
