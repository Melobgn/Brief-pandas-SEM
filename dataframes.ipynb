## <span style = 'color : orange'>exercice 23 </span>
Quel est le taux de croissance moyen en fonction de la zone géographique ?

```python
data_2018 = df[df['year'] == 2018]
data_2019 = df[df['year'] == 2019]
merged_data = pd.merge(data_2018, data_2019, on=['area_country','country','geo_area','item','type','food_supply_(Kcal/capital/day)','food_supply_quantity_(kg/capital/year)','fat_supply_quantity_(g/capital/day)','protein_supply_quantity_(g/capital/day)'] 
,suffixes=('_2018', '_2019'))
merged_data['le_taux_de_croissance_moyen'] = ((merged_data['pop_1000_hab_2019'] - merged_data['pop_1000_hab_2018']) / merged_data['pop_1000_hab_2018']) * 100
taux_croissance = merged_data 
taux_croissance = taux_croissance[['geo_area','le_taux_de_croissance_moyen']].groupby('geo_area').mean().round(2).reset_index()
taux_croissance = taux_croissance.sort_values(by='le_taux_de_croissance_moyen' , ascending=False)
taux_croissance.plot(kind='bar', x='geo_area', y='le_taux_de_croissance_moyen', color='orange')
taux_croissance.to_csv('le taux de croissance.csv')
taux_croissance
```

## <span style = 'color : orange'>exercice 24 </span>
 Calculez la disponibilité de nourriture totale par pays et par année, en kcal et kg de protéines. Attention aux unités de mesure !


```python
# Multiplication de la disponibilité alimentaire par habitant par jour par la population
df['food_supply_(Kcal/capital/day)'] = df['food_supply_(Kcal/capital/day)'] * df['pop_1000_hab']

# Agrégation pour obtenir la disponibilité totale de kilocalories par pays et par année
dispo_nourriture_totale_kcal = df.groupby(['country', 'year'])[['food_supply_(Kcal/capital/day)']].agg(sum)
dispo_nourriture_totale_kcal = dispo_nourriture_totale_kcal.reset_index()
dispo_nourriture_totale_kcal = dispo_nourriture_totale_kcal.sort_values(['year', 'food_supply_(Kcal/capital/day)'], ascending=False)

# Multiplication de la quantité de protéines par habitant par jour par 1000 pour obtenir des grammes
df['protein_supply_quantity_(g/capital/day)'] = df['protein_supply_quantity_(g/capital/day)'] * 1000

# Agrégation pour obtenir la disponibilité totale de protéines en grammes par pays et par année
dispo_nourriture_totale_prot = df.groupby(['country', 'year'])[['protein_supply_quantity_(g/capital/day)']].agg(sum)
dispo_nourriture_totale_prot = dispo_nourriture_totale_prot.reset_index()
dispo_nourriture_totale_prot = dispo_nourriture_totale_prot.sort_values(['year', 'protein_supply_quantity_(g/capital/day)'], ascending=False)
```

```python
# print(f'{dispo_nourriture_totale_kcal}\n\n{dispo_nourriture_totale_prot}')
```

```python
# Fusion des DataFrames dispo_nourriture_totale_kcal et dispo_nourriture_totale_prot
merged_df = pd.merge(dispo_nourriture_totale_kcal, dispo_nourriture_totale_prot, on=['country', 'year'])

# Calcul du ratio énergie/poids pour chaque produit et pays
merged_df['ratio_energie_poids'] = merged_df['food_supply_(Kcal/capital/day)'] / merged_df['protein_supply_quantity_(g/capital/day)']

# Remplacement des valeurs infinies par NaN et suppression des lignes avec des valeurs nulles
merged_df.replace([np.inf, -np.inf], np.nan, inplace=True)
merged_df.dropna(subset=['ratio_energie_poids'], inplace=True)

# Utilisez la colonne 'item' pour le regroupement
moyenne_ratio_par_aliment = merged_df.groupby('item')['ratio_energie_poids'].mean()

# Afficher les résultats
print(moyenne_ratio_par_aliment)
```
