```python
#Question 24 : Calculez la disponibilité de nourriture totale par pays et par année, en kcal et kg de protéines. Attention aux unités de mesure !

df_copy = df.copy()
df_copy['food_supply_kcal_day'] = df['food_supply_kcal_day'] * df['pop_1000_hab']
dispo_nourriture_totale_kcal = df_copy.groupby(['country', 'year'])[['food_supply_kcal_day']].agg(sum)
dispo_nourriture_totale_kcal = dispo_nourriture_totale_kcal.reset_index()
dispo_nourriture_totale_kcal = dispo_nourriture_totale_kcal.sort_values(['year', 'food_supply_kcal_day'], ascending=False)
print(dispo_nourriture_totale_kcal)
# dispo_nourriture_totale_kcal.to_csv('disponibilite_nourriture_totale_pays_annee_kcal.csv')



df['protein_supply_quantity_g'] = df['protein_supply_quantity_g'] * 1000
dispo_nourriture_totale_prot = df.groupby(['country', 'year'])[['protein_supply_quantity_g']].agg(sum)
dispo_nourriture_totale_prot = dispo_nourriture_totale_prot.reset_index()
dispo_nourriture_totale_prot = dispo_nourriture_totale_prot.sort_values(['year', 'protein_supply_quantity_g'], ascending=False)
dispo_nourriture_totale_prot = dispo_nourriture_totale_prot.rename(columns={'country' : 'pays', 'year' : 'année', 'protein_supply_quantity_g' : 'kg de protéines'})
dispo_nourriture_totale_prot
# dispo_nourriture_totale_prot.to_csv('disponibilité_nourriture_totale_pays_annee_kgproteines.csv')

```
