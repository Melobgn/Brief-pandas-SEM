```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Filtrer le DataFrame pour ne prendre que les données de l'année 2019
df_2019 = df[df['year'] == 2019]

# Calculer la diversité alimentaire par pays
diversite_df = df_2019.groupby(['country'])['item'].nunique().sort_values(ascending=False).head(20)

# Filtrer la disponibilité alimentaire pour les pays présents dans diversite_df
dispo_alimentaire = df_2019[df_2019['country'].isin(diversite_df.index)].groupby('country')['food_supply_(Kcal/capital/day)'].sum()

# Réorganiser les données de disponibilité alimentaire en fonction de la diversité
dispo_alimentaire = dispo_alimentaire.reindex(diversite_df.index).head(20)

# Créer la figure et les axes
fig, ax = plt.subplots(figsize=(12, 6))

# Largeur des barres
bar_width = 0.4

# Position des barres pour diversité
positions_diversite = np.arange(len(diversite_df))

# Tracer les barres pour diversité
ax.bar(positions_diversite, diversite_df, width=bar_width, color='blue', label='Diversité Alimentaire')

# Position des barres pour disponibilité alimentaire
positions_dispo = positions_diversite + bar_width

# Tracer les barres pour disponibilité alimentaire
ax.bar(positions_dispo, dispo_alimentaire, width=bar_width, color='green', label='Disponibilité Alimentaire (kcal)')

# Ajuster les étiquettes de l'axe x
ax.set_xticks(positions_dispo - bar_width / 2)
ax.set_xticklabels(diversite_df.index, rotation=50)

# Ajuster l'axe y pour commencer à 70
ax.set_ylim(70, max(diversite_df) + 5)

# Ajouter une légende
ax.legend(loc='upper left')

# Ajouter un titre
plt.title('Comparaison entre la Diversité Alimentaire et la Disponibilité Alimentaire (kcal) par Pays en 2019', fontsize=16)

# Affichage du graphique
plt.show()
```
